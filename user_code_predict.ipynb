```python
import pandas as pd
```

```python
df_user = pd.read_csv('./data/user_id_text.csv')
df_pre = pd.read_csv('./data/predict_output_text.csv')
```

```python
print(df_user)
print(df_pre)
```

```python
df_pre_mean = df_pre.groupby('node_index')['predict'].mean()
df_pre_mean = df_pre_mean.reset_index(drop = False)
df_pre_mean.columns = ['node_index','pre_val']
df_pre_mean
```

```python
df_pre_mean.pre_val.unique()
```

```python
def predict_code(df):
    #df.loc[(df['pre_val'] > 1.25) & (df['pre_val'] < 1.75), 'predict'] = 0
    #df.loc[df['pre_val'] <= 1.25, 'predict'] = 1
    #df.loc[df['pre_val'] >= 1.75, 'predict'] = 2
    df.loc[df['pre_val'] < 1.5, 'predict'] = 1
    df.loc[df['pre_val'] >= 1.5, 'predict'] = 2
    return df
```

```python
df_predict = predict_code(df_pre_mean)
df_predict
```

```python
df_user_predict = df_user.merge(df_predict, left_on = 'index', right_on = 'node_index', how = 'left')
df_user_predict
```
